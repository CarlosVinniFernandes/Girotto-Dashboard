```python
verificacao_valores_ausentes = df.isnull().sum()
verificacao_valores_ausentes
```

```python
Q1 = df['Numero_de_faltas'].quantile(0.25)
Q3 = df['Numero_de_faltas'].quantile(0.75)
IQR = Q3 - Q1


limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR


outliers = df[(df['Numero_de_faltas'] < limite_inferior) | (df['Numero_de_faltas'] > limite_superior)]

print("Outliers detectados:")
print(outliers)

sns.boxplot(x=df['Numero_de_faltas'])
plt.show()
```

```python
mediana = df['Numero_de_faltas'].median()
df['Numero_de_faltas'] = df['Numero_de_faltas'].apply(lambda x: mediana if x > limite_superior or x < limite_inferior else x)
```

```python
Q1 = df['Nota_final'].quantile(0.25)
Q3 = df['Nota_final'].quantile(0.75)
IQR = Q3 - Q1


limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR


outliers = df[(df['Nota_final'] < limite_inferior) | (df['Nota_final'] > limite_superior)]

print("Outliers detectados:")
print(outliers)

sns.boxplot(x=df['Nota_final'])
plt.show()
```

```python
mediana = df['Nota_final'].median()
df['Nota_final'] = df['Nota_final'].apply(lambda x: mediana if x > limite_superior or x < limite_inferior else x)
```

```python
matrix_de_correlacao = df.corr(numeric_only=True)


plt.figure(figsize=(10, 6))


sns.heatmap(matrix_de_correlacao, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)


plt.title("Matriz de Correlação entre Variáveis", fontsize=14)


plt.show()
```
